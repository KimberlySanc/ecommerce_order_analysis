```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df.isnull().sum()
```

```python
age_labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.qcut(df['customer_age'], q=4, labels = age_labels)
```

```python
df[['customer_age', 'age_group']].head(10)
```

```python
spend_labels = ['Low','Medium','High']
df['customer_lifetime_value_tier'] = pd.qcut(df['customer_lifetime_value'], q=3, labels = spend_labels)
```

```python
df[['customer_lifetime_value', 'customer_lifetime_value_tier']].head(10)
```

```python
df['gross_sales_amount'] = df['unit_price'] * df['quantity']
```

```python
df[['gross_sales_amount', 'unit_price','quantity' ]].head(10)
```
